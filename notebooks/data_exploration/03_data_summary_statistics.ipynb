```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='parenting_style', hue='parenting_style', palette='Set1', legend=False)
plt.title('Distribution of Parenting Styles')
plt.xlabel('Parenting Style')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 12))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 2, i+1)
    sns.boxplot(data=df, y=col)
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(18, 14))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 2, i+1)
    sns.violinplot(data=df, x='parenting_style', y=col, hue='parenting_style', palette='Set2', legend=False)
    plt.xticks(rotation=45)
    plt.title(f'{col} by Parenting Style')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 12))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 2, i+1)
    sns.histplot(df[col], kde=True, color='darkblue')
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```
